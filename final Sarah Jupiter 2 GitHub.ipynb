```python
# Import Libraries

import pandas as pd # For data manupulation and analysis
import numpy as np # For mathematical computation
```

```python
# Load data from the working directory and store it as an object called dataframe

dataframe = pd.read_table("cars-sample35.txt", header = None, sep = ",", 
                          names = ["price", "maintenance_cost", "numberofdoors", "NumberOfpassengers",
                                  "Luggage_capacity", "safety_rating", "ClassificationOfVehicle"]) 
```

```python
# View the content of the created dataframe with loaded data

dataframe
```

The loaded data saved as an data frame object called dataframe has 35 observation(records) and 7 features(variables). The data is in a tabular form with 35 rows and 7 colums.

## Find the "number of passengers" value for each auto mobile having a "price" value of "med". Create a new list to store your findings and be sure to print your results. 

```python
'''Create a subset of the dataframe that stores  the "number of passengers" value for each auto mobile having a 
"price" value of "med". The subset is stored in an object called med_auto'''

med_auto = dataframe.loc[: , ["price", "NumberOfpassengers"]].query('price == "med"') 
```

```python
# View the created subset dataframe with the autos having a price value of med

med_auto
```

```python
''' Create a new list to store  the "number of passengers" value for each auto having a "price" value of 
"med" and store the list named med_auto_list '''

med_auto_list = [med_auto.columns.values.tolist()] + med_auto.values.tolist()
```

```python
''' print the list that stores the "number of passengers" value for each auto having a "price" value of 
"med" '''

print(med_auto_list)
```

```python
''' print the list that stores the "number of passengers" value for each auto having a "price" value of 
"med" '''

print(med_auto.loc[:,"NumberOfpassengers"].values.tolist())
```

```python
''' print the list that stores the index value for each automobile having a price value of "high" and a 
maintenance value that is not "low" '''

print(high_auto_list)
```

## Create a new list containing the only the integer equivalents of the doors values. Keep in mind that the lists you have created thus far are composed solely of strings. If you find any values of '5more' in your list, convert them to a '5'. After converting the '5more' values to '5', convert all of the items in your list to their numeric equivalent and calculate the average number of doors across all 35 autos using whichever of Python's built in functions you require. Print your result. 

```python
''' Create a subset of the dataframe that stores the number of doors value. The subset is stored in an object
called number_of_doors '''

number_of_doors = dataframe.loc[: , ["numberofdoors"]]
```

```python
''' View the created subset dataframe with the number of doors values '''

number_of_doors
```

```python
''' View observations in the two_doors_auto subset with "5more" as value of numberofdoors column''' 

number_of_doors.loc[number_of_doors["numberofdoors"] == "5more", :]
```

```python
# Find any values of '5more' in numberofdoors column in number_of_doors subset, convert them to a '5'

number_of_doors.loc[number_of_doors["numberofdoors"] == "5more", :] = 5
```

```python
''' View the new subset dataframe with the number of doors values after "5more" have been converted to "5" '''

number_of_doors
```

```python
# Check data type of column numberofdoors in the subset number_of_doors

print(number_of_doors["numberofdoors"].dtypes)
```

```python
# Convert the data type of numberofdoors column to numeric equivalent

number_of_doors["numberofdoors"] = number_of_doors["numberofdoors"].astype(int)
```

```python
''' Create a new list to store the integer equivalents of the number of doors values and store the list named
as number_of_doors_list'''

number_of_doors_list = [number_of_doors.columns.values.tolist()] + number_of_doors.values.tolist()
```

```python
''' print the list that stores the integer equivalents of the number of doors values '''

print(number_of_doors_list)
```

```python
''' print the list that stores the integer equivalents of the number of doors values '''

print(number_of_doors.loc[:, "numberofdoors"].values.tolist())
```

```python
# calculate the average number of doors across all 35 autos and store it as average_number_of_doors

average_number_of_doors = np.mean(number_of_doors_list[1:])
```

```python
# print the average number of doors
print ("Average number of doors: ", average_number_of_doors)
```
